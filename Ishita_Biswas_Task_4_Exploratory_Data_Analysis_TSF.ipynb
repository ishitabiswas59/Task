```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
```

```python
df = pd.read_csv("/content/globalterrorismdb_0718dist.csv", encoding='latin-1')
df.head()
```

```python
df.columns.values
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1,inplace=True)
df.isnull().sum()
```

```python
df.corr()
```

```python
fig,axes = plt.subplots(1,1,figsize=(12,7))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(df.corr().iloc[:15,:15], annot=True)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x=df['iyear'], hue='success', data=df)
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Year')
plt.title('Increase in Terror',size=25, fontweight="bold")
plt.show()
```

```python
sns.countplot(x=df['region_txt'], hue='success', data=df)
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Region')
plt.show()
```
